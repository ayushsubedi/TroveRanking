```python
import pandas as pd
import plotly
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
import plotly.figure_factory as ff
init_notebook_mode(connected=True)
import os
```

```python
df = pd.DataFrame(columns=['date', 'name', 'count'])
for subdir, dirs, files in os.walk('datasources/'):
    for file in files:
        filepath = subdir + os.sep + file
        df = df.append(pd.read_csv(filepath))
```

```python
df_trove = df[df['name']=="https://app.trovenow.com"]
```

```python
df_trove
```

```python
iplot([go.Bar(x=df_trove.date, y=df_trove['count'])])
```

```python
iplot([go.Bar(x=df_trove.date, y=df_trove.index)])
```
